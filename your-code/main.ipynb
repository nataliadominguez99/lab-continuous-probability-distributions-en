```python
from scipy.stats import uniform

# Set parameters for the uniform distribution
low = 4
high = 20   
size = 10   

data= uniform.rvs(4, 20, size=100)

# Plot histogram
plt.hist(data, bins=20, color='pink', edgecolor='purple', density=True)
plt.title("Uniform Distribution between {} and {}".format(low, high))
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
#function

def plot_uniform_distribution(low, high, size):
    data = uniform.rvs(low, high - low, size=size)
    plt.hist(data, bins=20, color='pink', edgecolor='purple', density=True)
    plt.title("Uniform Distribution between {} and {}".format(low, high))
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.show()
```

## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
from scipy.stats import norm 
```

```python
# Define the mean and standard deviation
mean = 10              # Mean (center) of the distribution
std_deviation = 1      # Standard deviation (spread) of the distribution

# Define the number of samples to generate
data = 1000            # Number of data points to generate

# Generate synthetic data following a normal (Gaussian) distribution
# np.random.normal(loc, scale, size)
#   loc   = mean
#   scale = standard deviation
#   size  = number of values to generate
norm_distribution = np.random.normal(mean, std_deviation, data)

# Display the first 10 generated values (optional)
print(norm_distribution[:10])
```

```python
# Plot a histogram of generated data
plt.figure(figsize=(8, 6))  # Set the figure size
# Create histogram:
# - bins=70 creates 70 bars
# - density=True normalizes the histogram to form a probability density
# - alpha=0.6 sets transparency
# - color='b' sets bar color to blue
# - label adds a legend entry
plt.hist(norm_distribution, bins=70, density=True, alpha=0.6, color='b', label='Generated Data')

# Generate a range of x values for plotting the PDF curve
# np.linspace(start, stop, num_points) creates 1000 equally spaced values
x_values = np.linspace(min(norm_distribution), max(norm_distribution), 1000)

# Calculate the corresponding PDF (Probability Density Function) values
# norm.pdf(x, mean, std_dev) gives the height of the normal curve for each x
pdf_values = norm.pdf(x_values, mean, std_deviation)

# Plot the PDF as a smooth red curve
# 'r-' means red line, lw=2 means line width of 2
plt.plot(x_values, pdf_values, 'r-', lw=2, label='Normal Distribution PDF')

# Add descriptive labels and legend
plt.xlabel('Value')                    # Label for the x-axis
plt.ylabel('Probability Density')      # Label for the y-axis
plt.title('Normal Distribution (mean=10, std=1)')  # Plot title
plt.legend()                           # Show legend box

# Display the plot
plt.show()
```

```python
# Define the mean and standard deviation
mean = 10              # Mean (center) of the distribution
std_deviation = 50      # Standard deviation (spread) of the distribution

# Define the number of samples to generate
data = 1000            # Number of data points to generate

# Generate synthetic data following a normal (Gaussian) distribution
# np.random.normal(loc, scale, size)
#   loc   = mean
#   scale = standard deviation
#   size  = number of values to generate
norm_distribution = np.random.normal(mean, std_deviation, data)

# Display the first 10 generated values (optional)
print(norm_distribution[:10])
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

How are the two distributions different?

```python
def numbers_distribution(scale, size):
    data = np.random.exponential(scale=scale, size=size)
    plt.figure(figsize=(10, 5))
    plt.hist(data, bins=100, edgecolor='lightblue', color='purple')
    plt.title(f'Uniform Distribution: scale={scale}, size={size}')
    plt.xlabel('Mean')
    plt.ylabel('Count')
    plt.grid(True)
    plt.show()
numbers_distribution(10, 10) 
numbers_distribution(10,100)
```

The mean changes, so the distribution changes as well. 

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# Parameters
mean_time = 10  # minutes
lambda_param = 1 / mean_time
x = 15  # minutes

# Probability calculation
probability = 1 - math.exp(-lambda_param * x)

print(f"Probability that a customer spends less than 15 minutes: {probability:.4f}")
```
